```python
import pandas as pd
import numpy as np
```

```python
cond1_column = 'order_amount'
cond2_column = 'num_instalments_initial'
cond1 = 1300
cond2 = 18
violation_column_name = 'rule_violated'
```

```python
def apply_rules(df, violation_column_name, cond1_column: str, cond1: int, cond2_column: str, cond2: int):
    # Apply the rule: if amount > 1300, number_of_payments must be >= 18, otherwise reject (assign target = 0)
    df[violation_column_name] = np.where((df[cond1_column] > cond1) & (df[cond2_column] < cond2), 1, 0)
    df.loc[df[violation_column_name] == 1, 'target'] = 0  # Assuming rejected applications are non-fraudulent
    return df
```

```python
# load dataset
df = pd.read_pickle("prueba_tecnica_2023.pkl", compression="bz2")
# dimensionality
print(df.shape)
# check first rows of the dataframe
df.head()
```

```python
# apply custom rule
df = apply_rules(df,violation_column_name, cond1_column, cond1, cond2_column, cond2)
```
